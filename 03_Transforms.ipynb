##Transforms
- We use transforms to perfome some manipulation of the data and make it suitable for training.
- `transform`: to modify the features
- `target_transform`: to modify the labels

```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

# The FashionMNIST features are in PIL image format and the labels are integers.
# We need to normalize the features to tensors and the labels to one-hot encoded tensors.
ds = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    # ToTensor converts a PIL image into a FloatTensor and scales the image's pixel in the range [0., 1.]
    transform=ToTensor(),
    # Lambda transforms apply any user-defiend lambda function.
    target_transform=Lambda(lambda y: torch.zeros(10, dtype=torch.float).scatter_(
        dim=0, 
        index=torch.tensor(y), 
        value=1
    ))
)
```
